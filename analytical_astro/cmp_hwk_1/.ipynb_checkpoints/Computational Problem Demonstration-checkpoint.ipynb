# Problem 1: Convert Cartesian State to Keplerian and Analytically Propagate

## Part I

## Set initial conditions: 
$$ \mu = 4 \times 10^5 km^3/sec^2 $$ 
$$ \textbf{r_0} = [6 \times 10^3, 6 \times 10^3, 6 \times 10^3] km $$
$$ \textbf{v_0} = [-5,5,0] km/s $$


```python
pos = np.array([6*10**3, 6*10**3, 6*10**3]) # km 
vel = np.array([-5, 5, 0]) #km / sec
mu = 4*10**5 # km^3 / sec^2
time = 0 # sec
state = Cartesian_State(pos, vel, mu, time)
```

## Now we propagate the orbit over one period with a 60 second interval

```python
prop_1 = PropagatorKeplerian(state)
period = prop_1.initial_state.period
rad, vel, time = prop_1.propagate(period, 60)
```

## Now, lets plot this orbit in position and velocity space

```python
sns.lineplot(time, rad)
plt.title("Radius vs Time")
plt.xlabel("time since reference epoch (sec)")
plt.ylabel("Radius magnitude (km)")
```

```python
sns.lineplot(time, vel, color="orange")
plt.title("Velocity vs Time")
plt.xlabel("time since reference epoch (sec)")
plt.ylabel("Velocity magnitude (km/sec)")
```

# And now the position vs Velocity

```python
sns.lineplot(rad, vel)
plt.title("Radius vs Velocity")
plt.xlabel("Velocity magnitude (km/sec)")
plt.ylabel("Radius Magnitude (km/sec)")
```

# Now to plot in XYZ space 

```python
x_pos_1, y_pos_1, z_pos_1 = ([], [], [])
for state in prop_1.prop_states: 
    x, y, z = state.cart_pos
    x_pos_1.append(x)
    y_pos_1.append(y)
    z_pos_1.append(z)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(x_pos_1, y_pos_1, z_pos_1, 'blue')
plt.title("X-Y-Z space orbit plot")
```

## Part II 

Using the following as the initial conditions

```python
radius = 10000 # km 
raan = np.radians(45)
incl = np.radians(135)
arg_peri = np.radians(-90)
true_anom = 0 
mu = 4 * 10**5 # km /s
time = 0
```

We can get the periapsis velocity and angular momentum for each value of e using the following

```python
eccentricities = [0,.25,.5,.75,.99]
r_as = [(radius * -ecc - radius)/(ecc - 1) for ecc in eccentricities]
a_s = [(radius + ra)/2 for ra in r_as]
hs = [math.sqrt(mu * a * (1 - ecc**2)) for a, ecc in zip(a_s, eccentricities)]
```

Now we can construct our states and propagators using our keplerian state module and propagator module.

```python
states_list = [Keplerian_State(radius, h, incl, raan, ecc, arg_peri, true_anom, mu, time) for h, ecc 
               in zip(hs, eccentricities)]
prop_list = [PropagatorKeplerian(state) for state in states_list]
```

## Now let's plot the orbits for each value of e 

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

# plot the 0 eccentricity case
x_pos, y_pos, z_pos = ([], [], [])
for time in np.arange(60, period, 60):
    circ_state = states_list[0].prop_circular(time)
    x, y, z = circ_state.cart_pos
    x_pos.append(x)
    y_pos.append(y)
    z_pos.append(z)
ax.scatter(x_pos, y_pos, z_pos)

for prop in prop_list[1:]: 
    prop.propagate(5 * period, 60)
    x_pos, y_pos, z_pos = ([], [], [])
    for s in prop.prop_states: 
        x, y, z = s.cart_pos
        x_pos.append(x)
        y_pos.append(y)
        z_pos.append(z)
    ax.scatter(x_pos, y_pos, z_pos)

plt.title("X-Y-Z space orbit plot")
ax.legend(['e = 0', 'e = .25', 'e = .5', 'e = .75', 'e = .99'], loc="center left",)
```

# Problem 3: Numberical Integration of Cartesian State

First we set up the desired initial conditions (from problem 1) and then run the propagator

```python
pos_cart = np.array([6*10**3, 6*10**3, 6*10**3]) # km 
vel_cart = np.array([-5, 5, 0]) #km / sec
mu_cart = 4*10**5 # km^3 / sec^2
time_cart = 0 # sec
state_cart = Cartesian_State(pos_cart, vel_cart, mu_cart, time_cart)
```

```python
prop_cart = PropagatorCartesian(state_cart)
positions, velocities, times = prop_cart.propagate(period, 60)
```

Now we can plot this result!

```python
x_pos_cart, y_pos_cart, z_pos_cart = ([], [], [])
for pos in positions: 
    x_cart, y_cart, z_cart = pos
    x_pos_cart.append(x_cart)
    y_pos_cart.append(y_cart)
    z_pos_cart.append(z_cart)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(x_pos_cart, y_pos_cart, z_pos_cart, 'blue')
plt.title("X-Y-Z space orbit plot")
```

## Now let's plot this against the Keplerian propagation from part 1

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(x_pos_cart, y_pos_cart, z_pos_cart, 'blue')
ax.scatter3D(x_pos_1, y_pos_1, z_pos_1, 'orange')
ax.legend(['Numerical Integration', 'Analytical Orbit'], loc="center left",)
plt.title("Keplerian vs Numerical Integration X-Y-Z orbit plot")
```

We can see that the two answers are almost perfectly aligned
